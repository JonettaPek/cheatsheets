**Responsive Images (same content)**  
> <code>&lt;img src="url" 
    alt="description" 
    height="480" 
    width="360" 
    srcset="url-480 1x, url-960 2x, url-1440 3x, url-1920 4x"/&gt;</code>  
  
> <code>&lt;img src="url" 
    alt="description" 
    height="480" 
    width="360" 
    srcset="url-480 480w, url-960 960w, url-1440 1440w, url-1920 1920w" 
    sizes="(max width: 480px) 240px, (max width: 960px) 480px,(max width: 1440px) 960px,"/&gt;</code>  
- File formats: .jpg, .png, .gif, .svg
- Different formats compress the image differently
- Find a balance between file size and image quality  

- Aspect ratio (height by width: 4 by 3)  
- The images should have the same aspect ratio  

- srcset attribute lists a set of image files in which the browser decides which file to download and display considering the pixel density of the screen or viewport width, assuming the image will be 100% of the screen (disregarding the layout).  

- sizes attribute is a user-provided guideline on which image file to download and display based on the screen width, taking into account layout. E.g. The 240px image is used with a screen with a width between 0-480px.  

**Responsive Images (different content)**  
<code>&lt;picture&gt;
    &lt;source media="(min-width:600px)"
            srcset="../dog-320.jpg 320w,
                    ../dog-480.jpg 480w,
                    ../dog-720.jpg 720w,
                    ../dog-960.jpg 960w,
                    ../dog-1440.jpg 1440w,
                    ../dog-1920.jpg 1920w" /&gt;
    &lt;source srcset="../dog-cropped-320.jpg 320w,
                        ../dog-cropped-480.jpg 480w,
                        ../dog-cropped-720.jpg 720w,
                        ../dog-cropped-960.jpg 960w" /&gt;
    &lt;img src="url" 
        alt="description" 
        height="480" 
        width="360" 
 &lt;picture&gt;</code>  

- The `<picture>` element is incompatible with older browers like internet explorer 11
- The `<img>` element has to be used together with the `<picture>` element
- The `<source>` elements provide alternative options to image files

**Generic container**  
`<div>` is a block element  
`<span>` is an inline element

All global attributes are compatible with them for css or js targeting.
